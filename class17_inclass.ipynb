```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
# Generate the exponential data
values = np.random.exponential(3.45, 100)
values.sort()
months = range(1, 101)

# Create a DataFrame
data = pd.DataFrame({'Months': months, 'Values': values})

# scikit-learn Regression
X = data[['Months']]
y = data['Values']
```

```python
print(X)
print(y)
```

```python
# scikit-learn
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
```

```python
# Create the seaborn plot
sns.scatterplot(data=data, x='Months', y='Values', label="Data")
sns.regplot(data=data, x='Months', y='Values', color="red", scatter=False, label="Seaborn Regression", line_kws={"linewidth":5})
plt.plot(data["Months"], y_pred, label="Scikit-learn Regression", color="lightgreen")
plt.legend()
plt.show()
```
