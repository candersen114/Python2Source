```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/BriDeWaltCCC/PFDADataSets/main/Spotify_dataset.csv')
topdf = df[df['popularity'] > 95]
fig, ax = plt.subplots()
ax.bar(topdf['artists'], topdf['duration_ms'])
plt.xticks(rotation=30, ha="right")
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/BriDeWaltCCC/PFDADataSets/main/New_York_City_Community_Health_Survey.csv')
fig, ax = plt.subplots()
dfFiltered = df[df["Prevelance"].str.startswith("Prevalence")]
ax.plot(dfFiltered['Year'], dfFiltered['Binge Drinking'])
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/BriDeWaltCCC/PFDADataSets/main/Spotify_dataset.csv')
fig, ax = plt.subplots(nrows=2)

#ax.plot(df['energy'], df['speechiness'])
#ax.plot(df['energy'], df['tempo'], c='b')

# That graph was two line graphs that were completely overlapping each other.
#ax.scatter(df['energy'], df['speechiness'])
# That's a little bit better, but still nearly impossible to derive any meaning from it.
# I think what we really want is the average speechiness for each energy value; tempo will probably be similar
dfGrouped = df.groupby("energy")
mean_speechiness = dfGrouped["speechiness"].mean()
ax[0].plot(mean_speechiness.index, mean_speechiness)
# That at least gets us the trend

mean_tempo = dfGrouped["tempo"].mean()
ax[1].plot(mean_tempo.index, mean_tempo)

ax[1].set_xlabel("Energy")
ax[0].set_ylabel("Speechiness")
ax[1].set_ylabel("Tempo")
```
