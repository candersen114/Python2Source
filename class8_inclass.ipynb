```python
import locale
locale.setlocale(locale.LC_ALL, "en_US.UTF-8")

dfDates = pd.read_csv("https://raw.githubusercontent.com/BriDeWaltCCC/PFDADataSets/main/MockDataForFilteringandFixing.csv", dtype_backend="pyarrow")
dfDates["Date"] = dfDates["Date"].astype("date32[pyarrow]")
dfDates["DateEuroFormat"] = ""
dfDates["CostFloat"] = 0.0
print(dfDates.dtypes)

for i in dfDates.index:
    dfDates.loc[i, "DateEuroFormat"] = f"{dfDates.loc[i, "Date"]:%d/%m/%Y}"
    dfDates.loc[i, "CostFloat"] = locale.atof(dfDates.loc[i, "Cost"].strip(locale.localeconv()["currency_symbol"]))

display(dfDates)
```

```python
# Pat Lacey's version from class

# import the datetime module for the date formatting
from datetime import datetime

# read the data into a dataframe
url = 'https://raw.githubusercontent.com/BriDeWaltCCC/PFDADataSets/main/MockDataForFilteringandFixing.csv'
df = pd.read_csv(url)
print(df)

# start with one date string - first index
date = df.loc[0,'Date']
print(date)

# convert the string into a datetime date object
date_object = datetime.strptime(date, "%m/%d/%Y")

# convert the date object into a string with the correct format
new_date = date_object.strftime("%d/%m/%Y")
print(new_date)

# then we can do it with the whole column

# create a for loop based on the shape of the dataframe
for i in range(0, df.shape[0]):
    # the same processing code for the single string example
    date = df.loc[i,'Date']
    date_object = datetime.strptime(date, "%m/%d/%Y")
    new_date = date_object.strftime("%d/%m/%Y")
    # overwrite the data in the dataframe
    df.at[i, 'Date'] = new_date
print(df)

# there is another way to do this without a for loop
# the apply function uses a lambda x to reformat the date for every index
# more condensed code but same logic
# with this method, I changed the date format back to the initial format to show both methods working

df["Date"] = df["Date"].apply(
    lambda x: 
        datetime.strptime(x, "%d/%m/%Y").strftime("%m/%d/%Y")
)
df
```
