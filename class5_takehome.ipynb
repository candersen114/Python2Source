```python
# Imports - avoid importing multiple copies
import pandas as pd
```

```python
# Create a dataframe for tracking information about a product
# At least 5 rows and 5 columns
# Explain why columns and data types are chosen

dictComputers = {
    "SKU": ["BW1516512S", "BW1516512B", "BP1316512B", "BP1516512BL", "BP1532512BL", "BP15321TBL"],
    "Brand": ["BananaTech", "BananaTech", "BananaTech", "BananaTech", "BananaTech", "BananaTech"],
    "Model Line": ["BananaBook Wind", "BananaBook Wind", "BananaBook Pro", "BananaBook Pro", "BananaBook Pro", "BananaBook Pro"],
    "Model Name": ["BananaBook Wind 15-inch 16GB/512GB Silver", "BananaBook Wind 15-inch 16GB/512GB Black", "BananaBook Pro 13-inch 16GB/512GB Black", "BananaBook Pro 15-inch 16GB/512GB Blue", "BananaBook Pro 15-inch 32GB/512GB Blue", "BananaBook Pro 15-inch 32GB/1TB Blue"],
    "Screen Size (in)": [15.6, 15.6, 13.6, 15.6, 15.6, 15.6],
    "Storage Type": ["SSD", "SSD", "SSD", "SSD", "SSD", "SSD"],
    "Storage Size (GB)": [512, 512, 512, 512, 512, 1024],
    "Memory Type/Speed": ["DDR5-5600", "DDR5-5600", "DDR5-5600", "DDR5-5600", "DDR5-5600", "DDR5-5600"],
    "Memory Size (GB)": [16, 16, 16, 16, 32, 32],
    "Price": [1199.99, 1199.99, 1399.99, 1599.99, 1799.99, 1999.99]}

dfComputers = pd.DataFrame(data=dictComputers)
dfComputers = dfComputers.convert_dtypes(dtype_backend="pyarrow")
print(dfComputers.dtypes)
display(dfComputers)

print("DATA TYPES\n==========")
print("SKU, Brand, Model Line, Model Name, Storage Type, Memory Type/Speed: Strings (freeform text makes sense for these)")
print("Screen Size, Storage Size, Memory Size, Price: All numeric. Instead of storing screen size, for example, as 15.6 inches,")
print("    store it as numeric, because that allows comparisons between products. This in turn allows us to answer questions such as")
print("    \"What is the cheapest machine available with 32GB of memory?\", because for some users, that's the most important.")
print("    Screen Size and Price are doubles (non-integer), storage size and memory size are integers.")
```
