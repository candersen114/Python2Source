```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("ds_salaries.csv")
X_train, X_test, y_train, y_test = train_test_split(df[["salary"]], df["salary_in_usd"], test_size=.2, random_state=42)
#                                                   X - has to be a list of columns, not a single column
#                                                                   Y - single column
```

```python
# build our linear regression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# apply the model to our test data
y_output = model.predict(X_test)
display(y_output)
```

```python
# Calculate mean squared error and r-squared between the model's prediction and the actual Y values in our test set
mse = mean_squared_error(y_test, y_output)
r2 = r2_score(y_test, y_output)

print(mse)
print(r2)
```

```python
plt.scatter(df["salary"], df["salary_in_usd"])
plt.plot(X_test, y_output, color="blue", linewidth=3)
```

# Exercise 2
- Create a graph from one of the two we just went over (bar or histogram). 
- Make sure to change at least 5 properties of the graph.
- Use the DataFrame below for data

```python
df = pd.read_csv('https://raw.githubusercontent.com/BriDeWaltCCC/PFDADataSets/main/ds_salaries.csv')
display(df)
print(df.dtypes)
```

```python
# Sort the DataFrame by the X-axis column
df_sorted = df.sort_values(by='salary_in_usd', ascending=True)
# create a figure and axes objects
fig, ax = plt.subplots()
# plot the 'x' column as a histogram
ax.hist(df_sorted['salary_in_usd'], rwidth=0.9, color="purple")
# add a label to the x-axis
ax.set_xlabel( 'Salary Range')
ax.set_ylabel("Frequency")
# add a title to the plot
ax.set_title( 'Histogram of Salary Range' )
# Set the X axis labels at an angle
plt.xticks(rotation=45, ha='right')
# display the plot
plt.show()
```

# Take-Home Exercise
Create a line plot using Matplotlib that displays the number of fruit sold by a fruit stand over the course of a week. The fruit stand sold apples, bananas, and oranges.

```python
dictFruit = {
    "Day": ["1", "2", "3", "4", "5"],
    "Apples": [10, 12, 15, 14, 11],
    "Bananas": [20, 22, 25, 24, 21],
    "Oranges": [15, 17, 20, 19, 16]
}

dfFruit = pd.DataFrame(data=dictFruit)

fig, ax = plt.subplots()
ax.plot(dfFruit['Day'], dfFruit['Apples'], "-x", color="red", label="Apples")
ax.plot(dfFruit['Day'], dfFruit['Bananas'], "-o", color="gold", label="Bananas")
ax.plot(dfFruit['Day'], dfFruit['Oranges'], "-*", color="orange", label="Oranges")
# add labels to the x and y axes
ax.set_xlabel('Day')
ax.set_ylabel('Fruit Sold')
# Add title
fig.suptitle("Fruit sold per day")
fig.legend(title="Fruits", loc="center left", bbox_to_anchor=(1, 0.5))
# display the plot
plt.show()
```
